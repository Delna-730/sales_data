Profitability Analysis

```python
df = df[df['Revenue'] != 0].copy()
```

```python
df['Profit_Margin'] = (df['Profit'] / df['Revenue']) * 100
```

```python
df.groupby('Country')['Profit_Margin'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(df['Revenue'])
plt.show()
```

```python
Q1 = df['Revenue'].quantile(0.25)
Q3 = df['Revenue'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[
    (df['Revenue'] < lower) |
    (df['Revenue'] > upper)
]
```

```python
outliers[[
    'Product',
    'Product_Category',
    'Order_Quantity',
    'Unit_Price',
    'Revenue',
    'Profit',
    'Country'
]]
```
